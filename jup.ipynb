```python
import csv
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime

years = dict()
reader = csv.DictReader(open("./PlaneCrashes.csv", 'rt', encoding='utf-8'))
for row in reader:
    datum = datetime.strptime(row['Date'], '%Y-%m-%d %H:%M:%S')
    if datum.year not in years:
        years[datum.year] = 1
    else:
        years[datum.year] += 1

x = list(years.keys())
y = list(years.values())

plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Number of Entries per Year')
plt.grid(True)
plt.show()
```

```python
razlogi = dict()
reader = csv.DictReader(open('PlaneCrashes.csv', 'rt', encoding='utf-8'))
for row in reader:
    razlog = row['Crash cause']
    if razlog not in razlogi:
       razlogi[razlog] = 1
    else:
        razlogi[razlog] += 1

sorted_razlogi = dict(sorted(razlogi.items(), key=lambda item: item[1], reverse=True))

x = list(sorted_razlogi.keys())
y = list(sorted_razlogi.values())

plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.xlabel('Cause')
plt.ylabel('Count')
plt.title('Number of Entries per Cause')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

types = {}
reader = csv.DictReader(open('PlaneCrashes.csv', 'rt', encoding='utf-8'))
for row in reader:
    flight_type = row['Flight type']
    if flight_type.strip():  # Check if flight type is not empty
        types[flight_type] = types.get(flight_type, 0) + 1
    else:
        types['NA'] = types.get('NA', 0) + 1  # Handling missing values

sorted_types = dict(sorted(types.items(), key=lambda item: item[1], reverse=True))

x = list(sorted_types.keys())
y = list(sorted_types.values())

plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.xlabel('Flight Type')
plt.ylabel('Count')
plt.title('Number of Entries per Flight Type')
plt.xticks(rotation=90)

# Add 'NA' label under x-axis if it doesn't exist in the flight types
if 'NA' not in x:
    x.append('NA')
    y.append(0)  # Add 0 count for 'NA' flight type
    plt.xticks(range(len(x)), x)

plt.grid(True)
plt.show()
```
